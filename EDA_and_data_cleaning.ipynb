# DATA CLEANING

```python
# Clean 'Restaurant Name' column
dataframe['Restaurant Name']=dataframe['Restaurant Name'].str.strip('_,�')
dataframe['Restaurant Name']=dataframe['Restaurant Name'].str.replace('�','')
dataframe['Restaurant Name']=dataframe['Restaurant Name'].str.replace('_','')
dataframe['Restaurant Name']=dataframe['Restaurant Name'].str.replace('۱','')

print(dataframe['Restaurant Name'])
```

```python
# Clean 'City' column
dataframe['City']=dataframe['City'].str.strip('_,�')
dataframe['City']=dataframe['City'].str.replace('�','')
dataframe['City']=dataframe['City'].str.replace('_','')

print(dataframe['City'])
```

```python
# Clean 'Address' column
dataframe['Address']=dataframe['Address'].str.strip('_,�')
dataframe['Address']=dataframe['Address'].str.replace('�','')
dataframe['Address']=dataframe['Address'].str.replace('_','')

print(dataframe['Address'])
```

```python
# Clean 'Locality' column
dataframe['Locality']=dataframe['Locality'].str.strip('_,�')
dataframe['Locality']=dataframe['Locality'].str.replace('�','')
dataframe['Locality']=dataframe['Locality'].str.replace('_','')
dataframe['Locality']=dataframe['Locality'].str.replace('۱','')

print(dataframe['Locality'])
```

```python
# Clean 'Locality Verbose' column
dataframe['Locality Verbose']=dataframe['Locality Verbose'].str.strip('_,�')
dataframe['Locality Verbose']=dataframe['Locality Verbose'].str.replace('�','')
dataframe['Locality Verbose']=dataframe['Locality Verbose'].str.replace('_','')
dataframe['Locality Verbose']=dataframe['Locality Verbose'].str.replace('۱','')

print(dataframe['Locality Verbose'])
```

```python
 # Clean 'Currency' column
dataframe['Currency']=dataframe['Currency'].str.replace('�','£')

print(dataframe['Currency'])
```

```python
# Clean 'Cuisines' column
dataframe['Cuisines']=dataframe['Cuisines'].str.strip('_,�')
dataframe['Cuisines']=dataframe['Cuisines'].str.replace('�','')
dataframe['Cuisines']=dataframe['Cuisines'].str.replace('_','')
dataframe['Cuisines']=dataframe['Cuisines'].str.replace('۱','')

print(dataframe['Cuisines'])
```

```python
# Handel missing values (null values) in cuisines column
dataframe['Cuisines'] = dataframe['Cuisines']. fillna('')
```
